# One-Hot Encoding

```python
import numpy as np
import pandas as pd
```

```python
data = {'Categories': np.random.choice(['red', 'green', 'blue'], 100)}
```

```python
df = pd.DataFrame(data)
df.head(10)
```

```python
encoded_df = pd.get_dummies(df, dtype=int)
encoded_df.head(10)
```

```python
encoded_df2 = pd.get_dummies(df, drop_first=True, dtype=int)
encoded_df2.head(10)
```
